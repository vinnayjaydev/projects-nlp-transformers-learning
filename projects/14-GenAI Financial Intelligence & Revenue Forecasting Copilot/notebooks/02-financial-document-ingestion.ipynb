```python
# Start coding here
# ==========================================
# Notebook 02
# Financial Document Ingestion
# ==========================================

import pandas as pd
import numpy as np
import re

import nltk

from nltk.tokenize import sent_tokenize

nltk.download("punkt")
```

```python
financial_df = pd.read_csv("../data/financial_corporate_dataset.csv")

financial_df.head()
```

```python
financial_df[["quarter", "earnings_call", "risk_factors", "mda_section"]]
```

```python
def clean_financial_text(text):

    text = str(text)

    text = re.sub(r"\s+", " ", text)

    text = re.sub(r"[^a-zA-Z0-9.,%$ ]", "", text)

    text = text.strip()

    return text
```

```python
sample_text = financial_df.iloc[0]["combined_text"]

clean_financial_text(sample_text)
```

```python
financial_df["clean_text"] = financial_df["combined_text"].apply(clean_financial_text)
```

```python
financial_df[["quarter", "clean_text"]]
```

```python
financial_df["character_count"] = financial_df["clean_text"].apply(len)

financial_df[["quarter", "character_count"]]
```

```python
def sentence_splitter(text):

    return sent_tokenize(text)
```

```python
sentence_splitter(financial_df.iloc[0]["clean_text"])
```

```python
financial_df["sentences"] = financial_df["clean_text"].apply(sentence_splitter)
```

```python
financial_df[["quarter", "sentences"]]
```

```python
def create_chunks(text, chunk_size=3):

    sentences = sent_tokenize(text)

    chunks = []

    for i in range(0, len(sentences), chunk_size):

        chunk = " ".join(sentences[i : i + chunk_size])

        chunks.append(chunk)

    return chunks
```

```python
create_chunks(financial_df.iloc[0]["clean_text"])
```

```python
financial_df["chunks"] = financial_df["clean_text"].apply(create_chunks)
```

```python
financial_df[["quarter", "chunks"]]
```

```python
chunk_records = []

for _, row in financial_df.iterrows():

    for idx, chunk in enumerate(row["chunks"]):

        chunk_records.append(
            {
                "ticker": row["ticker"],
                "quarter": row["quarter"],
                "chunk_id": idx,
                "text": chunk,
            }
        )
```

```python
chunks_df = pd.DataFrame(chunk_records)
```

```python
chunks_df.head()
```

```python
chunks_df["text_length"] = chunks_df["text"].apply(len)
```

```python
chunks_df.head()
```
