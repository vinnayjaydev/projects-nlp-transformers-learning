```python
# ==========================================
# Notebook 11
# Multimodal Feature Fusion
# ==========================================

import pandas as pd
import numpy as np
```

```python
financial_df = pd.read_csv("../data/financial_macro_dataset.csv")

financial_df.head()
```

```python
print("Rows:", len(financial_df))

print("Columns:", len(financial_df.columns))

financial_df.columns
```

```python
numerical_features = [
    "gross_margin_pct",
    "operating_income_million",
    "net_income_million",
    "eps",
]
```

```python
text_features = [
    "weighted_sentiment",
    "guidance_score",
    "positive_guidance_score",
    "total_risk_score",
    "normalized_risk_score",
    "macro_risk_score",
]
```

```python
required_columns = numerical_features + text_features

for col in required_columns:

    print(col, "->", col in financial_df.columns)
```

```python
numerical_matrix = financial_df[numerical_features]

numerical_matrix
```

```python
text_matrix = financial_df[text_features]

text_matrix
```

```python
numerical_matrix.shape
```

```python
text_matrix.shape
```

```python
multimodal_features = pd.concat([numerical_matrix, text_matrix], axis=1)
```

```python
multimodal_features
```

```python
multimodal_features.shape
```

```python
target = financial_df["revenue_million"]
```

```python
multimodal_features.describe()
```

```python
multimodal_features.isnull().sum()
```

```python
multimodal_features = multimodal_features.fillna(0)
```

```python
correlation_matrix = multimodal_features.corr()

correlation_matrix
```

```python
correlation_with_revenue = (
    pd.concat([multimodal_features, target.rename("revenue_million")], axis=1)
    .corr()["revenue_million"]
    .sort_values(ascending=False)
)

correlation_with_revenue
```

```python
correlation_with_revenue.to_frame("correlation")
```

```python
modeling_df = multimodal_features.copy()

modeling_df["revenue_million"] = target
```

```python
modeling_df.head()
```

```python
multimodal_features.to_csv("../data/multimodal_feature_matrix.csv", index=False)
```

```python
modeling_df.to_csv("../data/multimodal_forecasting_dataset.csv", index=False)
```

```python
correlation_with_revenue.to_csv("../data/feature_revenue_correlations.csv")
```
