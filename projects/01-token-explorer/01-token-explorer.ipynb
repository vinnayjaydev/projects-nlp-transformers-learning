```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace

# Create empty BPE tokenizer
tokenizer = Tokenizer(BPE())

# Split on whitespace before learning merges
tokenizer.pre_tokenizer = Whitespace()

# Configure trainer
trainer = BpeTrainer(
    vocab_size=1000,
    min_frequency=2,
    special_tokens=["[PAD]", "[UNK]", "[CLS]", "[SEP]", "[MASK]"],
)

# Train tokenizer
files = [
    r"D:\Books\projects-nlp-transformers-learning\projects\01-token-explorer\combined_corpus.txt"
]

tokenizer.train(files, trainer)

# Save tokenizer
tokenizer.save("tokenizer.json")

print("Tokenizer training completed!")
```

```python
from tokenizers import Tokenizer
from tokenizers.models import WordPiece
from tokenizers.pre_tokenizers import Whitespace
from tokenizers.trainers import WordPieceTrainer

tokenizer = Tokenizer(WordPiece(unk_token="[UNK]"))

tokenizer.pre_tokenizer = Whitespace()

trainer = WordPieceTrainer(
    vocab_size=1000, special_tokens=["[PAD]", "[UNK]", "[CLS]", "[SEP]", "[MASK]"]
)

path = r"D:\Books\projects-nlp-transformers-learning\projects\01-token-explorer\combined_corpus.txt"


tokenizer.train([path], trainer)

tokenizer.save("wordpiece_tokenizer.json")
```

```python
models = {
    "BERT (WordPiece)": "bert-base-uncased",
    "GPT2 (BPE)": "gpt2",
    "T5 (SentencePiece)": "t5-small",
}

sample_text = "Understanding tokenization is essential for LLMs."

results = []

for name, model in models.items():
    tok = AutoTokenizer.from_pretrained(model)
    tokens = tok.tokenize(sample_text)

    results.append(
        {"Tokenizer": name, "Number of Tokens": len(tokens), "Tokens": tokens}
    )

comparison_df = pd.DataFrame(results)

comparison_df
```
