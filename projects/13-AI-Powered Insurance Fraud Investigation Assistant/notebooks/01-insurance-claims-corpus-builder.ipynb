```python
# ====================================
# Notebook 01
# Insurance Claims Corpus Builder
# ====================================

import pandas as pd
import numpy as np

from faker import Faker

import random
```

```python
fake = Faker()

random.seed(42)
np.random.seed(42)
```

```python
mechanic_shops = [
    "Rapid Auto Repair",
    "Elite Collision Center",
    "Metro Auto Works",
    "Prime Vehicle Repair",
    "Trusted Auto Garage",
]
```

```python
clinics = [
    "Care First Clinic",
    "Advanced Spine Center",
    "Metro Injury Clinic",
    "Health Plus Medical",
    "Wellness Recovery Center",
]
```

```python
lawyers = [
    "Anderson Legal",
    "Justice Partners",
    "Smith & Associates",
    "Premier Injury Law",
    "Accident Recovery Group",
]
```

```python
normal_claim_statements = [
    "I was driving through an intersection when another vehicle struck my car.",
    "A vehicle rear-ended me while I was waiting at a traffic signal.",
    "The driver failed to yield and collided with my vehicle.",
    "I was traveling within the speed limit when the accident occurred.",
    "The vehicle changed lanes unexpectedly and hit my car.",
]
```

```python
police_reports = [
    "Police determined the claimant was not at fault.",
    "Witnesses confirmed another driver caused the accident.",
    "Traffic camera footage supports claimant statement.",
    "Police observed damage consistent with reported collision.",
    "Accident report indicates claimant followed traffic rules.",
]
```

```python
def create_claim(claim_id):

    return {
        "claim_id": claim_id,
        "claimant_name": fake.name(),
        "vehicle_vin": fake.bothify(text="???###########"),
        "mechanic_shop": random.choice(mechanic_shops),
        "clinic_name": random.choice(clinics),
        "lawyer": random.choice(lawyers),
        "claimant_statement": random.choice(normal_claim_statements),
        "police_report": random.choice(police_reports),
        "adjuster_notes": fake.sentence(),
        "medical_bill": random.randint(500, 25000),
        "fraud_label": 0,
    }
```

```python
claims = []
```

```python
for i in range(1, 16):

    claims.append(create_claim(f"CLM{i:03}"))
```

```python
claims_df = pd.DataFrame(claims)
```

```python
claims_df.head(11)
```

```python
claims_df.loc[2, "claimant_statement"] = (
    "I was stopped at a red light when another vehicle hit me."
)
```

```python
claims_df.loc[2, "fraud_label"] = 1
```

```python
template_text = """

My neck and lower back experienced severe pain immediately after impact.

The collision caused significant emotional distress and loss of mobility.

I continue to experience discomfort.
"""
```

```python
claims_df.loc[5, "claimant_statement"] = template_text
```

```python
claims_df.loc[10, "claimant_statement"] = template_text
```

```python
claims_df.loc[[5, 10], "fraud_label"] = 1
```

```python
ring_shop = "Rapid Auto Repair"

ring_clinic = "Metro Injury Clinic"

ring_lawyer = "Premier Injury Law"
```

```python
ring_claims = [7, 8, 9]
```

```python
for idx in ring_claims:

    claims_df.loc[idx, "mechanic_shop"] = ring_shop

    claims_df.loc[idx, "clinic_name"] = ring_clinic

    claims_df.loc[idx, "lawyer"] = ring_lawyer

    claims_df.loc[idx, "fraud_label"] = 1
```
