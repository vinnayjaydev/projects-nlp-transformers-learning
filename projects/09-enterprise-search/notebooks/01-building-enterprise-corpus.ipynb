```python
# ==========================================================
# Notebook 01: Building Enterprise Corpus
# ==========================================================

import random
import uuid

import numpy as np
import pandas as pd
```

```python
# -----------------------------------
# Configuration
# -----------------------------------

RANDOM_SEED = 42

np.random.seed(RANDOM_SEED)

random.seed(RANDOM_SEED)
```

```python
sources = ["Slack", "Confluence", "Jira"]

departments = ["Engineering", "HR", "Finance", "Legal"]
```

```python
slack_messages = [
    {
        "source": "Slack",
        "title": "Phoenix Deployment Discussion",
        "content": (
            "The Phoenix-2026 deployment "
            "is scheduled for next Friday. "
            "Database migration will happen "
            "during the maintenance window."
        ),
        "author": "Alice",
    },
    {
        "source": "Slack",
        "title": "Model Training Update",
        "content": (
            "The semantic search model "
            "has been retrained using "
            "new support ticket data."
        ),
        "author": "Bob",
    },
]

slack_messages
```

```python
confluence_pages = [
    {
        "source": "Confluence",
        "title": "Phoenix Architecture Wiki",
        "content": (
            "Project Phoenix-2026 follows "
            "a microservice architecture "
            "using Kubernetes and FastAPI."
        ),
        "author": "Charlie",
    },
    {
        "source": "Confluence",
        "title": "Enterprise Search Design",
        "content": (
            "Hybrid search combines BM25 "
            "keyword retrieval and "
            "semantic vector retrieval."
        ),
        "author": "Diana",
    },
]

confluence_pages
```

```python
jira_tickets = [
    {
        "source": "Jira",
        "title": "PHX-245 Database Migration",
        "content": (
            "Complete PostgreSQL migration " "before Phoenix production launch."
        ),
        "author": "Eric",
    },
    {
        "source": "Jira",
        "title": "SEC-104 ACL Enhancement",
        "content": (
            "Implement metadata-based "
            "access control filtering "
            "for enterprise search."
        ),
        "author": "Fiona",
    },
]

jira_tickets
```

```python
enterprise_records = slack_messages + confluence_pages + jira_tickets
```

```python
enterprise_records
```

```python
for idx, record in enumerate(enterprise_records, start=1):

    record["doc_id"] = f"DOC{idx:04d}"

    record["department"] = random.choice(departments)

    record["created_date"] = f"2026-01-{10 + idx}"

    record["tags"] = random.sample(
        [
            "enterprise-search",
            "phoenix",
            "database",
            "acl",
            "fastapi",
            "semantic-search",
            "kubernetes",
        ],
        k=2,
    )
```

```python
enterprise_df = pd.DataFrame(enterprise_records)

enterprise_df
```

```python
print("Total Documents:", len(enterprise_df))

print("\nSources:")

print(enterprise_df["source"].value_counts())

print("\nDepartments:")

print(enterprise_df["department"].value_counts())
```

```python
source_summary = (
    enterprise_df.groupby("source").size().reset_index(name="document_count")
)

source_summary
```

```python
OUTPUT_PATH = "data/processed/"

import os

os.makedirs(OUTPUT_PATH, exist_ok=True)

enterprise_df.to_csv(os.path.join(OUTPUT_PATH, "enterprise_corpus.csv"), index=False)

print("Enterprise corpus saved!")
```
