```python
documents = {
    "data/hr/Benefits_2026.txt": hr_doc,
    "data/finance/Annual_Report_2026.txt": finance_doc,
    "data/engineering/AI_Roadmap.txt": engineering_doc,
}

for filepath, content in documents.items():

    with open(filepath, "w", encoding="utf-8") as file:

        file.write(content)

print("Documents saved.")
```

```python
document_schema = {
    "document_id": "",
    "filename": "",
    "department": "",
    "doc_type": "",
    "created_at": "",
    "content": "",
}

document_schema
```

```python
def load_document(filepath):

    with open(filepath, "r", encoding="utf-8") as file:

        text = file.read()

    return text
```

```python
def create_document_record(filepath):

    content = load_document(filepath)

    department = Path(filepath).parent.name

    return {
        "document_id": str(uuid.uuid4()),
        "filename": Path(filepath).name,
        "department": department,
        "doc_type": "corporate_document",
        "created_at": "2026-01-01",
        "content": content,
    }
```

```python
records = []

for root, dirs, files in os.walk("data"):

    for file in files:

        filepath = os.path.join(root, file)

        record = create_document_record(filepath)

        records.append(record)
```

```python
documents_df = pd.DataFrame(records)

documents_df.head()
```

```python
documents_df["char_count"] = documents_df["content"].apply(len)

documents_df[["filename", "department", "char_count"]]
```

```python
repository = {}

for _, row in documents_df.iterrows():

    repository[row["document_id"]] = {
        "filename": row["filename"],
        "department": row["department"],
        "content": row["content"],
    }
```
