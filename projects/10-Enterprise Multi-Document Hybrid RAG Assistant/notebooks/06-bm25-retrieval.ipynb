```python
# =====================================================
# Notebook 06
# BM25 Retrieval
# =====================================================

import pandas as pd

from rank_bm25 import BM25Okapi
```

```python
children_df = pd.read_csv("artifacts/child_chunks.csv")

children_df.head()
```

```python
children_df.iloc[0]["content"]
```

```python
tokenized_corpus = []

for text in children_df["content"]:

    tokens = text.lower().split()

    tokenized_corpus.append(tokens)
```

```python
tokenized_corpus[0]
```

```python
bm25 = BM25Okapi(tokenized_corpus)
```

```python
def bm25_search(query, top_k=5):

    query_tokens = query.lower().split()

    scores = bm25.get_scores(query_tokens)

    results = children_df.copy()

    results["bm25_score"] = scores

    results = results.sort_values(by="bm25_score", ascending=False)

    return results.head(top_k)
```

```python
bm25_search("health insurance")
```

```python
bm25_search("revenue")
```

```python
metadata_df = pd.read_csv("artifacts/document_metadata.csv")
```

```python
children_df = children_df.merge(
    metadata_df[["document_id", "department"]], on="document_id", how="left"
)
```

```python
department_bm25 = {}

department_docs = {}
```

```python
for department in children_df["department"].unique():

    subset = children_df[children_df["department"] == department].reset_index(drop=True)

    corpus = [text.lower().split() for text in subset["content"]]

    department_bm25[department] = BM25Okapi(corpus)

    department_docs[department] = subset
```

```python
department_bm25.keys()
```

```python
def bm25_department_search(query, department, top_k=5):

    bm25_index = department_bm25[department]

    docs = department_docs[department]

    query_tokens = query.lower().split()

    scores = bm25_index.get_scores(query_tokens)

    docs = docs.copy()

    docs["bm25_score"] = scores

    docs = docs.sort_values(by="bm25_score", ascending=False)

    return docs.head(top_k)
```

```python
bm25_department_search("health insurance", "hr")
```

```python
bm25_department_search("revenue growth", "finance")
```
