```python
from sentence_transformers import SentenceTransformer

from sklearn.metrics.pairwise import cosine_similarity

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

sentence1 = "How can I learn AI?"
sentence2 = "Beginner guide to Artificial Intelligence."

embeddings = model.encode([sentence1, sentence2])

score = cosine_similarity([embeddings[0]], [embeddings[1]])

print("Bi-Encoder Similarity:", round(score[0][0], 3))
```

```python
from sentence_transformers import CrossEncoder

cross_model = CrossEncoder("cross-encoder/stsb-roberta-base")

pairs = [["How can I learn AI?", "Beginner guide to Artificial Intelligence."]]

score = cross_model.predict(pairs)

print("Cross-Encoder Score:", round(float(score[0]), 3))
```
