```python
# ==========================================================
# Notebook 2: Similarity Metrics
# Cosine Similarity, Dot Product,
# Euclidean Distance & Manhattan Distance
# ==========================================================

print("Notebook 2: Similarity Metrics")
```

```python
import numpy as np

A = np.array([2, 3])
B = np.array([4, 6])
C = np.array([-2, 1])

print("Sentence A Vector:", A)
print("Sentence B Vector:", B)
print("Sentence C Vector:", C)
```

```python
A = np.array([2, 3])
B = np.array([4, 6])

dot_product = A[0] * B[0] + A[1] * B[1]

print("Dot Product =", dot_product)
```

```python
dot_product = np.dot(A, B)

print("NumPy Dot Product =", dot_product)
```

```python
D = np.array([40, 60])

print("A·B =", np.dot(A, B))
print("A·D =", np.dot(A, D))
```

```python
dot_product = np.dot(A, B)

norm_A = np.linalg.norm(A)
norm_B = np.linalg.norm(B)

cosine_similarity = dot_product / (norm_A * norm_B)

print("Dot Product :", dot_product)
print("||A|| :", round(norm_A, 3))
print("||B|| :", round(norm_B, 3))
print("Cosine Similarity :", round(cosine_similarity, 3))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

A = np.array([[2, 3]])
B = np.array([[4, 6]])

score = cosine_similarity(A, B)

print(score)
```

```python
pairs = [
    (np.array([2, 3]), np.array([4, 6])),
    (np.array([1, 0]), np.array([0, 1])),
    (np.array([1, 1]), np.array([-1, -1])),
]

for idx, (x, y) in enumerate(pairs, 1):

    score = np.dot(x, y) / (np.linalg.norm(x) * np.linalg.norm(y))

    print(f"Example {idx}: {round(score,3)}")
```

```python
euclidean = np.linalg.norm(A - B)

print("Euclidean Distance:", round(euclidean, 3))
```

```python
manhattan = np.sum(np.abs(A - B))

print("Manhattan Distance:", manhattan)
```

```python
A = np.array([2, 3])
B = np.array([4, 6])

dot = np.dot(A, B)

cosine = dot / (np.linalg.norm(A) * np.linalg.norm(B))

euclidean = np.linalg.norm(A - B)

manhattan = np.sum(np.abs(A - B))

print(f"Dot Product         : {dot:.3f}")
print(f"Cosine Similarity   : {cosine:.3f}")
print(f"Euclidean Distance  : {euclidean:.3f}")
print(f"Manhattan Distance  : {manhattan:.3f}")
```

```python
A = np.array([0.5, 0.8])
B = np.array([0.6, 0.9])
C = np.array([-0.5, -0.3])


def cosine(x, y):
    return np.dot(x, y) / (np.linalg.norm(x) * np.linalg.norm(y))


print("A vs B:", round(cosine(A, B), 3))

print("A vs C:", round(cosine(A, C), 3))

print("B vs C:", round(cosine(B, C), 3))
```
