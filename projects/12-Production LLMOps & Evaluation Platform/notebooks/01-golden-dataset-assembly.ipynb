```python
# Start coding here
# ============================================
# Notebook 01
# Golden Dataset Assembly
# ============================================

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
```

```python
evaluation_columns = ["question", "contexts", "generated_response", "ground_truth"]

evaluation_columns
```

```python
evaluation_data = [
    {
        "question": "What is AWS?",
        "contexts": "AWS stands for Amazon Web Services. It is a cloud computing platform.",
        "generated_response": "AWS is Amazon's cloud platform.",
        "ground_truth": "AWS is Amazon Web Services, a cloud computing platform.",
    },
    {
        "question": "What is Azure?",
        "contexts": "Microsoft Azure is a cloud platform offered by Microsoft.",
        "generated_response": "Azure is Microsoft's cloud service.",
        "ground_truth": "Microsoft Azure is a cloud computing platform offered by Microsoft.",
    },
    {
        "question": "What is Google Cloud?",
        "contexts": "Google Cloud Platform provides cloud computing services.",
        "generated_response": "Google Cloud offers cloud services.",
        "ground_truth": "Google Cloud Platform provides cloud computing services.",
    },
]
```

```python
evaluation_df = pd.DataFrame(evaluation_data)

evaluation_df
```

```python
evaluation_df.shape
```

```python
def validate_schema(dataframe):

    required_columns = ["question", "contexts", "generated_response", "ground_truth"]

    missing_columns = [col for col in required_columns if col not in dataframe.columns]

    if len(missing_columns) > 0:

        raise ValueError(f"Missing columns: {missing_columns}")

    print("Schema validation passed")
```

```python
validate_schema(evaluation_df)
```

```python
evaluation_df.isnull().sum()
```

```python
evaluation_df.describe(include="object")
```

```python
evaluation_df["question_length"] = evaluation_df["question"].str.len()
```

```python
evaluation_df["answer_length"] = evaluation_df["ground_truth"].str.len()
```

```python
evaluation_df[["question_length", "answer_length"]].describe()
```

```python
topics = ["AWS", "Azure", "Google Cloud", "Oracle Cloud", "IBM Cloud"]
```

```python
synthetic_data = []

for topic in topics:

    synthetic_data.append(
        {
            "question": f"What is {topic}?",
            "contexts": f"{topic} is a cloud platform.",
            "generated_response": f"{topic} provides cloud services.",
            "ground_truth": f"{topic} is a cloud computing platform.",
        }
    )
```

```python
synthetic_df = pd.DataFrame(synthetic_data)

synthetic_df.head()
```

```python
evaluation_df = pd.concat([evaluation_df, synthetic_df], ignore_index=True)
```

```python
evaluation_df.shape
```

```python
while len(evaluation_df) < 20:

    row = {
        "question": f"Question {len(evaluation_df)}",
        "contexts": "Sample context",
        "generated_response": "Sample generated response",
        "ground_truth": "Sample ground truth",
    }

    evaluation_df = pd.concat([evaluation_df, pd.DataFrame([row])], ignore_index=True)
```

```python
evaluation_df.shape
```

```python
evaluation_df.head()
```

```python
train_df, test_df = train_test_split(evaluation_df, test_size=0.2, random_state=42)
```

```python
print("Train:", len(train_df))

print("Test:", len(test_df))
```

```python
def dataset_health_report(df):

    report = {
        "rows": len(df),
        "columns": len(df.columns),
        "null_values": df.isnull().sum().sum(),
        "avg_question_length": round(df["question"].str.len().mean(), 2),
    }

    return report
```
