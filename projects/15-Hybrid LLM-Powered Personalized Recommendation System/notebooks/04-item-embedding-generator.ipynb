```python
# ==========================================
# Notebook 04
# Item Embedding Generator
# ==========================================

import pandas as pd
import numpy as np

from sentence_transformers import SentenceTransformer
```

```python
profiles_df = pd.read_csv("../data/item_profiles.csv")

profiles_df.head()
```

```python
print("Total Profiles:", len(profiles_df))

print()

profiles_df.columns
```

```python
embedding_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
embedding_model
```

```python
documents = profiles_df["content_profile"].tolist()
```

```python
len(documents)
```

```python
documents[0]
```

```python
item_embeddings = embedding_model.encode(documents, show_progress_bar=True)
```

```python
item_embeddings.shape
```

```python
item_embeddings[0]
```

```python
len(item_embeddings[0])
```

```python
embedding_df = pd.DataFrame(item_embeddings)

embedding_df.head()
```

```python
embedding_df.insert(0, "item_id", profiles_df["item_id"])
```

```python
embedding_df.head()
```

```python
embedding_df.isnull().sum().sum()
```

```python
embedding_df.describe()
```

```python
item_a = item_embeddings[0]

item_b = item_embeddings[1]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(item_a.reshape(1, -1), item_b.reshape(1, -1))[0][0]

similarity
```

```python
similarity_matrix = cosine_similarity(item_embeddings)

similarity_matrix
```

```python
similarity_matrix.shape
```

```python
similarity_df = pd.DataFrame(
    similarity_matrix, index=profiles_df["title"], columns=profiles_df["title"]
)

similarity_df
```

```python
target_index = 0

scores = similarity_matrix[target_index]

scores
```

```python
ranked_idx = np.argsort(scores)[::-1]
```

```python
for idx in ranked_idx[:5]:

    print()

    print(profiles_df.iloc[idx]["title"])

    print(round(scores[idx], 4))
```
