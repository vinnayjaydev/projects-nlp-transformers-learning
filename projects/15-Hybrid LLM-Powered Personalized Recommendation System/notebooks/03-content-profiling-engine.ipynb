```python
# ==========================================
# Notebook 03
# Content Profiling Engine
# ==========================================

import pandas as pd
import numpy as np
```

```python
items_df = pd.read_csv("../data/items.csv")

items_df.head()
```

```python
print("Total Items:", len(items_df))

print()

items_df.columns
```

```python
items_df.iloc[0]
```

```python
tag_mapping = {
    "Startup": "startup fintech founders entrepreneurship venture-capital",
    "Artificial Intelligence": "ai machine-learning llm deep-learning transformers",
    "Technology": "cloud computing distributed-systems scalability",
    "Data Engineering": "etl pipelines analytics warehousing",
    "Machine Learning": "machine-learning predictive-modeling ai",
    "Health": "health nutrition wellness",
    "Wellness": "habits productivity self-improvement",
}
```

```python
items_df["tags"] = items_df["category"].map(tag_mapping)

items_df.head()
```

```python
items_df["content_profile"] = (
    "Title: "
    + items_df["title"]
    + " Category: "
    + items_df["category"]
    + " Tags: "
    + items_df["tags"]
    + " Description: "
    + items_df["description"]
)
```

```python
items_df[["item_id", "content_profile"]].head()
```

```python
print(items_df.iloc[0]["content_profile"])
```

```python
items_df["profile_length"] = items_df["content_profile"].apply(len)
```

```python
items_df[["title", "profile_length"]]
```

```python
items_df["profile_length"].describe()
```

```python
items_df.sort_values(by="profile_length", ascending=False)[["title", "profile_length"]]
```

```python
items_df.groupby("category")["profile_length"].mean()
```

```python
recommendation_corpus = items_df["content_profile"].tolist()
```
