```python
# =====================================================
# Notebook 10
# Graph Memory & State Compression
# =====================================================

from typing import TypedDict
from typing import List
from typing import Dict

import numpy as np
import pandas as pd

from sentence_transformers import SentenceTransformer

from transformers import pipeline
```

```python
llm = pipeline(
    "text-generation", model="Qwen/Qwen2.5-1.5B-Instruct", max_new_tokens=256
)
```

```python
llm = pipeline(
    "text-generation", model="Qwen/Qwen2.5-1.5B-Instruct", max_new_tokens=256
)
```

```python
class EnterpriseResearchState(TypedDict):

    research_goal: str

    raw_findings: List[str]

    compressed_memory: str

    memory_archive: List[str]

    final_report: str

    iteration_count: int

    status: str

    metadata: Dict
```

```python
state = {
    "research_goal": "Compare cloud costs",
    "raw_findings": [
        "AWS pricing increased 8%",
        "Azure pricing remained stable",
        "Google Cloud storage reduced 4%",
        "Oracle Cloud expanded GPU offerings",
        "IBM Cloud remains niche player",
    ],
    "compressed_memory": "",
    "memory_archive": [],
    "final_report": "",
    "iteration_count": 1,
    "status": "",
    "metadata": {},
}
```

```python
def compress_findings_llm(findings):

    text = "\n".join(findings)

    prompt = f"""
Summarize these findings.

Keep:
- Important facts
- Trends
- Risks

Findings:

{text}
"""

    result = llm(prompt)

    return result[0]["generated_text"]
```

```python
summary = compress_findings_llm(state["raw_findings"])

print(summary)
```

```python
def memory_compression_agent(state):

    print("\nMemory Compression Agent Running...")

    summary = compress_findings_llm(state["raw_findings"])

    state["compressed_memory"] = summary

    state["status"] = "memory_compressed"

    return state
```

```python
updated_state = memory_compression_agent(state)

print(updated_state["compressed_memory"])
```

```python
def archive_memory(state):

    state["memory_archive"].append(state["compressed_memory"])

    return state
```

```python
updated_state = archive_memory(updated_state)

updated_state["memory_archive"]
```

```python
from sentence_transformers import SentenceTransformer

embedding_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

print("Embedding dimension:", embedding_model.get_sentence_embedding_dimension())
```

```python
def retrieve_memory(query, top_k=3):

    query_embedding = embedding_model.encode([query])

    distances, indices = memory_index.search(query_embedding.astype("float32"), top_k)

    results = []

    for idx in indices[0]:

        if idx < len(memory_texts):

            results.append(memory_texts[idx])

    return results
```

```python
def memory_manager_agent(state):

    if len(state["raw_findings"]) > 5:

        state = memory_compression_agent(state)

        state = archive_memory(state)

    return state
```

```python
memory_manager_agent(state)
```

```python
from langgraph.graph import StateGraph, END
```

```python
graph = StateGraph(EnterpriseResearchState)
```

```python
graph.add_node("memory_manager", memory_manager_agent)
```

```python
graph.set_entry_point("memory_manager")

graph.add_edge("memory_manager", END)
```

```python
app = graph.compile()
```

```python
result = app.invoke(state)
```
