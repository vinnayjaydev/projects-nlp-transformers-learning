```python
class ResearchState(TypedDict):

    research_goal: str

    queries_executed: List[str]

    raw_findings: List[str]

    deduplicated_findings: List[str]

    conflicts: List[str]

    critic_score: float

    draft_report: str

    final_report: str

    iteration_count: int

    status: str
```

```python
initial_state = {
    "research_goal": "Compare cloud costs against competitors",
    "queries_executed": [],
    "raw_findings": [],
    "deduplicated_findings": [],
    "conflicts": [],
    "critic_score": 0.0,
    "draft_report": "",
    "final_report": "",
    "iteration_count": 0,
    "status": "initialized",
}
```

```python
initial_state
```

```python
def researcher_agent(state):

    state["queries_executed"].append("cloud pricing comparison")

    state["raw_findings"].append("AWS pricing increased 8%")

    state["raw_findings"].append("Azure pricing remained stable")

    state["status"] = "research_completed"

    return state
```

```python
updated_state = researcher_agent(initial_state)

updated_state
```

```python
def deduplicator_agent(state):

    unique_findings = list(set(state["raw_findings"]))

    state["deduplicated_findings"] = unique_findings

    state["status"] = "deduplicated"

    return state
```

```python
updated_state = deduplicator_agent(updated_state)

updated_state["deduplicated_findings"]
```

```python
def critic_agent(state):

    findings = len(state["deduplicated_findings"])

    score = min(findings / 5, 1.0)

    state["critic_score"] = score

    state["status"] = "critic_completed"

    return state
```

```python
updated_state = critic_agent(updated_state)

updated_state["critic_score"]
```

```python
def writer_agent(state):

    report = "\n".join(state["deduplicated_findings"])

    state["draft_report"] = report

    state["status"] = "draft_created"

    return state
```

```python
updated_state = writer_agent(updated_state)

print(updated_state["draft_report"])
```

```python
def increment_iteration(state):

    state["iteration_count"] += 1

    return state
```

```python
updated_state = increment_iteration(updated_state)

updated_state["iteration_count"]
```

```python
state_history = []
```

```python
def save_snapshot(state):

    state_history.append(state.copy())
```

```python
save_snapshot(updated_state)

len(state_history)
```
