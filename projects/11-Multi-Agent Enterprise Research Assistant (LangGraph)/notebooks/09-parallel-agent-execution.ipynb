# 09 Parallel Agent Execution

```python
# ==================================================
# Notebook 09
# Parallel Agent Execution
# ==================================================

from typing import TypedDict
from typing import List
from typing import Dict

from langgraph.graph import StateGraph, END
```

```python
class ResearchState(TypedDict):

    research_goal: str

    finance_findings: List[str]

    competitor_findings: List[str]

    risk_findings: List[str]

    compliance_findings: List[str]

    merged_findings: List[str]

    final_report: str

    status: str

    metadata: Dict
```

```python
initial_state = {
    "research_goal": "Analyze cloud infrastructure strategy",
    "finance_findings": [],
    "competitor_findings": [],
    "risk_findings": [],
    "compliance_findings": [],
    "merged_findings": [],
    "final_report": "",
    "status": "",
    "metadata": {},
}
```

```python
def finance_agent(state):

    print("Finance Agent Running")

    state["finance_findings"] = [
        "Cloud spending increased 12%",
        "Infrastructure ROI improved 8%",
    ]

    return state
```

```python
def competitor_agent(state):

    print("Competitor Agent Running")

    state["competitor_findings"] = [
        "AWS expanded AI services",
        "Azure invested in GPU clusters",
        "Google Cloud reduced storage costs",
    ]

    return state
```

```python
def risk_agent(state):

    print("Risk Agent Running")

    state["risk_findings"] = [
        "Vendor lock-in risk identified",
        "Disaster recovery plan incomplete",
    ]

    return state
```

```python
def compliance_agent(state):

    print("Compliance Agent Running")

    state["compliance_findings"] = ["SOC2 controls verified", "GDPR review required"]

    return state
```

```python
def merge_agent(state):

    print("Merging Results")

    merged = []

    merged.extend(state["finance_findings"])

    merged.extend(state["competitor_findings"])

    merged.extend(state["risk_findings"])

    merged.extend(state["compliance_findings"])

    state["merged_findings"] = merged

    return state
```

```python
test_state = finance_agent(initial_state.copy())

test_state = competitor_agent(test_state)

test_state = risk_agent(test_state)

test_state = compliance_agent(test_state)

merge_agent(test_state)

test_state["merged_findings"]
```

```python
def writer_agent(state):

    report = "\n".join(state["merged_findings"])

    state["final_report"] = report

    state["status"] = "completed"

    return state
```

```python
graph = StateGraph(ResearchState)
```

```python
graph.add_node("finance", finance_agent)

graph.add_node("competitor", competitor_agent)

graph.add_node("risk", risk_agent)

graph.add_node("compliance", compliance_agent)

graph.add_node("merge", merge_agent)

graph.add_node("writer", writer_agent)
```

```python
def router_agent(state):

    print("Routing Research")

    return state
```

```python
graph.add_node("router", router_agent)
```

```python
graph.set_entry_point("router")
```

```python
graph.add_edge("router", "finance")

graph.add_edge("router", "competitor")

graph.add_edge("router", "risk")

graph.add_edge("router", "compliance")
```

```python
graph.add_edge("finance", "merge")

graph.add_edge("competitor", "merge")

graph.add_edge("risk", "merge")

graph.add_edge("compliance", "merge")
```

```python
graph.add_edge("merge", "writer")
```

```python
graph.add_edge("writer", END)
```

```python
app = graph.compile()
```
